```python
import os
import pandas as pd
import wandb
import matplotlib.pyplot as plt

api = wandb.Api()

runs = api.runs("aarnesh-ironman/ICML_LOES_6_HF_Models_Final")

rows = []

for run in runs:
    cfg = run.config

    dataset = cfg.get("dataset")
    model = cfg.get("model")
    seed = cfg.get("seed")

    if dataset is None or model is None or seed is None:
        continue

    history = run.history(pandas=True)

    if history.empty:
        continue

    x_col = None
    if "epoch" in history.columns:
        x_col = "epoch"
    elif "_step" in history.columns:
        x_col = "_step"
    else:
        continue

    if "val_acc" not in history.columns:
        continue

    history = history[[x_col, "val_acc"]].dropna()

    if history.empty:
        continue

    history[x_col] = history[x_col].astype(int)
    history["val_acc"] = history["val_acc"].astype(float)

    history["dataset"] = dataset
    history["model"] = model
    history["seed"] = str(seed)
    history["x"] = history[x_col]

    rows.append(history[["dataset", "model", "seed", "x", "val_acc"]])

df = pd.concat(rows, ignore_index=True)
```

```python
phase_lookup = {}

for run in runs:
    cfg = run.config

    dataset = cfg.get("dataset")
    model = cfg.get("model")
    seed = cfg.get("seed")
    phase = cfg.get("phase")

    if dataset is None or model is None or seed is None or phase is None:
        continue

    key = (dataset, model, str(seed))
    phase_lookup[key] = phase
```

```python
phase_lookup
```

```python
df["phase"] = df.apply(
    lambda r: phase_lookup.get((r["dataset"], r["model"], r["seed"])),
    axis=1
)

df = df.dropna(subset=["phase"])
```

```python
def normalize_phase(p):
    if p is None:
        return None
    p = str(p).lower()
    if "greedy" in p or "loes" in p:
        return "loes"
    if "last" in p:
        return "last"
    return None

df["phase_norm"] = df["phase"].apply(normalize_phase)
df = df.dropna(subset=["phase_norm"])
```

```python
df['phase_norm'].unique
```

```python
import os
import matplotlib.pyplot as plt

BASE_DIR = "epoch_vs_acc_per_seed"
os.makedirs(BASE_DIR, exist_ok=True)

for (dataset, model, seed), g in df.groupby(["dataset", "model", "seed"]):

    phases_present = set(g["phase"])
    if not {"greedy_loes_top3", "last_layer_baseline"}.issubset(phases_present):
        continue

    out_dir = os.path.join(BASE_DIR, dataset, model.replace("/", "_"))
    os.makedirs(out_dir, exist_ok=True)

    plt.figure(figsize=(6, 4))

    g_loes = g[g["phase"] == "greedy_loes_top3"].sort_values("x")
    g_last = g[g["phase"] == "last_layer_baseline"].sort_values("x")

    plt.plot(
        g_loes["x"],
        g_loes["val_acc"],
        label="LOES (k=3)",
        linestyle="-"
    )

    plt.plot(
        g_last["x"],
        g_last["val_acc"],
        label="Last Layer",
        linestyle="--"
    )

    plt.xlabel("Epoch / Step")
    plt.ylabel("Validation Accuracy")
    plt.title(f"{dataset} | {model} | seed {seed}")
    plt.legend()
    plt.tight_layout()

    plt.savefig(
        os.path.join(out_dir, f"seed_{seed}.png"),
        dpi=200
    )
    plt.close()
```

```python
import os
import pandas as pd
import wandb
import matplotlib.pyplot as plt
from tqdm import tqdm

api = wandb.Api()
runs = api.runs("aarnesh-ironman/ICML_LOES_6_HF_Models_Final")

TARGET_SEED = "42"

rows = []

print(f"[INFO] Target seed = {TARGET_SEED}")
print(f"[INFO] Total runs fetched = {len(runs)}")

for run in tqdm(runs, desc="Processing runs"):
    cfg = run.config

    dataset = cfg.get("dataset")
    model = cfg.get("model")
    seed = cfg.get("seed")
    phase = cfg.get("phase")

    if dataset is None or model is None or seed is None:
        print(f"[SKIP] Missing dataset/model/seed in run {run.name}")
        continue

    if str(seed) != TARGET_SEED:
        continue

    print(f"\n[RUN] {run.name}")
    print(f"      dataset={dataset}, model={model}, seed={seed}")

    if phase is None:
        name = run.name.lower()
        if "greedy" in name or "loes" in name:
            phase = "greedy_loes_top3"
            print("      phase inferred = LOES")
        elif "last" in name:
            phase = "last_layer_baseline"
            print("      phase inferred = LAST")
        else:
            print("      [SKIP] Could not infer phase")
            continue
    else:
        print(f"      phase from config = {phase}")

    history = run.history(pandas=True)
    if history.empty:
        print("      [SKIP] Empty history")
        continue

    if "epoch" in history.columns:
        x_col = "epoch"
    elif "_step" in history.columns:
        x_col = "_step"
    else:
        print("      [SKIP] No epoch or step column")
        continue

    if "val_acc" not in history.columns:
        print("      [SKIP] No val_acc in history")
        continue

    history = history[[x_col, "val_acc"]].dropna()
    if history.empty:
        print("      [SKIP] History empty after dropna")
        continue

    history[x_col] = history[x_col].astype(int)
    history["val_acc"] = history["val_acc"].astype(float)

    print(f"      [OK] Collected {len(history)} points")

    for _, r in history.iterrows():
        rows.append({
            "dataset": dataset,
            "model": model,
            "seed": str(seed),
            "phase": phase,
            "x": int(r[x_col]),
            "val_acc": float(r["val_acc"])
        })

df = pd.DataFrame(rows)

print("\n[INFO] Final DF shape:", df.shape)
print(df.groupby(["dataset", "model"])["phase"].unique())
```

```python
def normalize_phase(p):
    p = p.lower()
    if "greedy" in p or "loes" in p:
        return "loes"
    if "last" in p:
        return "last"
    return None

df["phase_norm"] = df["phase"].apply(normalize_phase)
df = df.dropna(subset=["phase_norm"])

print("\n[INFO] After phase normalization:")
print(df.groupby(["dataset", "model"])["phase_norm"].unique())

BASE_DIR = f"epoch_vs_acc_seed_{TARGET_SEED}"
os.makedirs(BASE_DIR, exist_ok=True)

for (dataset, model), g in tqdm(
    df.groupby(["dataset", "model"]),
    desc="Saving plots"
):

    phases = set(g["phase_norm"])
    print(f"\n[PLOT] {dataset} | {model} | phases={phases}")

    if phases != {"loes", "last"}:
        print("       [SKIP] Missing LOES or LAST")
        continue

    out_dir = os.path.join(BASE_DIR, dataset)
    os.makedirs(out_dir, exist_ok=True)

    plt.figure(figsize=(6, 4))

    g_loes = g[g["phase_norm"] == "loes"].sort_values("x")
    g_last = g[g["phase_norm"] == "last"].sort_values("x")

    plt.plot(g_loes["x"], g_loes["val_acc"], label="LOES (k=3)")
    plt.plot(g_last["x"], g_last["val_acc"], label="Last Layer", linestyle="--")

    plt.xlabel("Epoch / Step")
    plt.ylabel("Validation Accuracy")
    plt.title(f"{dataset} | {model} | seed {TARGET_SEED}")
    plt.legend()
    plt.tight_layout()

    fname = model.replace("/", "_") + ".png"
    plt.savefig(os.path.join(out_dir, fname), dpi=200)
    plt.close()

    print(f"       [SAVED] {fname}")
```
